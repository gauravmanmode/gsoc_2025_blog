```python
import numpy as np
import optimagic as om
results = {}
for algo in om.algorithms.AVAILABLE_ALGORITHMS.keys():
    if algo.startswith("nevergrad"):
        res = om.minimize(
            fun = lambda x: (x-2)@(x-2),
            params = np.ones(3),
            algorithm = algo
        )
        results[algo]=res


fig = om.criterion_plot(
    results,
    max_evaluations=500,
    monotone=True,
)
fig.show()
```
